```python
from sklearn.preprocessing import MinMaxScaler

categories = ['A', 'B', 'C', 'D', 'E', 'F']
X_test_normalized_all = {}
y_test_normalized_all = {}

X_validation_normalized_all = {}
y_validation_normalized_all = {}

X_train_normalized_all = {}
y_train_normalized_all = {}

# Calculate means and standard deviations before normalization
for category in categories:
    
    X_test_normalized_all[category] = []
    y_test_normalized_all[category] = []

    X_validation_normalized_all[category] = []
    y_validation_normalized_all[category] = []

    X_train_normalized_all[category] = []
    y_train_normalized_all[category] = []

    mean_before_X_train = X_train[category].mean(axis=0)
    std_before_X_train = X_train[category].std(axis=0)

    mean_before_X_validation = X_validation[category].mean(axis=0)
    std_before_X_validation = X_validation[category].std(axis=0)

    mean_before_X_test = X_test[category].mean(axis=0)
    std_before_X_test = X_test[category].std(axis=0)

    mean_before_y_train = y_train[category].mean(axis=0)
    std_before_y_train = y_train[category].std(axis=0)

    mean_before_y_validation = y_validation[category].mean(axis=0)
    std_before_y_validation = y_validation[category].std(axis=0)

    mean_before_y_test = y_test[category].mean(axis=0)
    std_before_y_test = y_test[category].std(axis=0)
    
    scalerx = MinMaxScaler()
    scalery = MinMaxScaler()

    X_train_normalized = scalerx.fit_transform(X_train[category])
    X_validation_normalized = scalerx.transform(X_validation[category])
    X_test_normalized = scalerx.transform(X_test[category])

    y_train_normalized = scalery.fit_transform(X_train[category])
    y_validation_normalized = scalery.transform(X_validation[category])
    y_test_normalized = scalery.transform(X_test[category])

    mean_after_X_train = X_train_normalized.mean(axis=0)
    std_after_X_train = X_train_normalized.std(axis=0)

    mean_after_X_validation = X_validation_normalized.mean(axis=0)
    std_after_X_validation = X_validation_normalized.std(axis=0)

    mean_after_X_test = X_test_normalized.mean(axis=0)
    std_after_X_test = X_test_normalized.std(axis=0)
 
    mean_after_y_train = y_train_normalized.mean(axis=0)
    std_after_y_train = y_train_normalized.std(axis=0)

    mean_after_y_validation = y_validation_normalized.mean(axis=0)
    std_after_y_validation = y_validation_normalized.std(axis=0)

    mean_after_y_test = y_test_normalized.mean(axis=0)
    std_after_y_test = y_test_normalized.std(axis=0)

    X_test_normalized_all[category] = np.concatenate(X_test_normalized)
    y_test_normalized_all[category] = np.concatenate(y_test_normalized)

    X_validation_normalized_all[category] = np.concatenate(X_validation_normalized)
    y_validation_normalized_all[category] = np.concatenate(y_validation_normalized)

    X_train_normalized_all[category] = np.concatenate(X_train_normalized)
    y_train_normalized_all[category] = np.concatenate(y_train_normalized)

    
    np.savez(f'normalized_data/normalization_params_{category}.npz',
            mean_before_X_test=mean_before_X_test,
            std_before_X_test=std_before_X_test,
            mean_after_X_test=mean_after_X_test,
            std_after_X_test=std_after_X_test,
            mean_before_X_validation=mean_before_X_validation,
            std_before_X_validation=std_before_X_validation,
            mean_after_X_validation=mean_after_X_validation,
            std_after_X_validation=std_after_X_validation,
            mean_before_X_train=mean_before_X_train,
            std_before_X_train=std_before_X_train,
            mean_after_X_train=mean_after_X_train,
            std_after_X_train=std_after_X_train,
            mean_before_y_test=mean_before_y_test,
            std_before_y_test=std_before_y_test,
            mean_after_y_test=mean_after_y_test,
            std_after_y_test=std_after_y_test,
            mean_before_y_validation=mean_before_y_validation,
            std_before_y_validation=std_before_y_validation,
            mean_after_y_validation=mean_after_y_validation,
            std_after_y_validation=std_after_y_validation,
            mean_before_y_train=mean_before_y_train,
            std_before_y_train=std_before_y_train,
            mean_after_y_train=mean_after_y_train,
            std_after_y_train=std_after_y_train)
             
```
